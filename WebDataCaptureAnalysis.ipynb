```python
import os
import re
import pandas as pd
import glob
from datetime import datetime
import warnings
warnings.filterwarnings("ignore")
```

```python
FileList = glob.glob("access_log_*95")
print(FileList)
```

```python
df_gen = (pd.read_csv(FileName, sep="!", encoding="ISO-8859-1", header=None) for FileName in FileList)
initialDataFrame = pd.concat(df_gen, ignore_index = True)
initialDataFrame.columns = ["Values"]
```

```python
#Extracting all hostnames from given logfile
HostRegex = r"(^\S+)\s"
Hostnames = [re.search(HostRegex, record).group(1)
               if re.search(HostRegex, record)
               else "No match found"
               for record in initialDataFrame["Values"]]
```

```python
Hostnames
```

```python
#Extracting all dates from given log file
dateRegex = r"(\d{2}[\/]\w*[\/]\d{4}:\d{2}:\d{2}:\d{2})"

dates = [re.search(dateRegex, record).group(1)
             if re.search(dateRegex, record)
             else "No match found"
             for record in initialDataFrame["Values"]]
```

```python
dates
```

```python
#Extracting all StatusCodes from the given Logfile
statusRegex = r"\s(\d{3})\s"

statuscodes = [re.search(statusRegex, record).group(1)
            if re.search(statusRegex, record)
            else "No match found"
            for record in initialDataFrame["Values"]]
```

```python
statuscodes
```

```python
#Extracting Bytes from the given logfile
bytesRegex = r"\s(\d+)$"
ListofBytes = [re.search(bytesRegex, record).group(1)
             if re.search(bytesRegex, record)
             else "No match found"
             for record in initialDataFrame["Values"]]
```

```python
ListofBytes
```

```python
initialDataFrame["Host"] = initialDataFrame["Values"].str.extract(HostRegex, expand = True)
initialDataFrame["Date"] = initialDataFrame["Values"].str.extract(dateRegex, expand = True)
initialDataFrame["Status"] = initialDataFrame["Values"].str.extract(statusRegex, expand = True)
initialDataFrame["Bytes"] = initialDataFrame["Values"].str.extract(bytesRegex, expand = True)
```

```python
#requiredDataFrame is a new dataframe which is created for having only the required fields namely Host,Date,StatusCode,Bytes
requiredDataFrame = initialDataFrame.filter(['Host','Date','Status','Bytes'], axis=1)
```

```python
#filling all null values with 0 as placeholder
requiredDataFrame["Bytes"].fillna(0, inplace = True)
```

```python
#All null values handled by placing 0 as placeholder
requiredDataFrame.isnull().any()
```

```python
#Converting Extracted Date to Required  YYYY-MM-DD HH:MM:SS
requiredDataFrame['Date'] = pd.to_datetime(requiredDataFrame['Date'], format='%d/%b/%Y:%H:%M:%S')
EarliestDate = requiredDataFrame.Date.min()
LatestDate = requiredDataFrame.Date.max()
print("Earliest Date is:",EarliestDate)
print("Latest Date is:",LatestDate)
```
